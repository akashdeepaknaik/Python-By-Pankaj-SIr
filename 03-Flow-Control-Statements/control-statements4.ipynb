## Pattern Printing using loop

```python
for row in range(1,6):
   for col in range(1,row+1):
      print(col,end='')
   print()
```

```python
for row in range(1,6):
   for col in range(1,row+1):
      print("*",end='')
   print()
```

```python
n=int(input("Enter a number :"))
for row in range(1,n+1):
   for col in range(1,row+1):
      print(col,end=" ")
   print()
```

```python
for row in range(1,5):
   print(" "*(4-row),end="")
   print("*"*(2*row-1))
print()
```

```python
n=int(input("Enter a number :"))
for row in range(1,n+1):
   print(" "*(n-row),end="")
   print("*"*(2*row-1))
print()
```

## loop with else

```python
## loop with else only there in python
for i in range(1,11):
   if i%50==0:
      break
   print(i)
else:
   print("Hello") #This will also execute 
```

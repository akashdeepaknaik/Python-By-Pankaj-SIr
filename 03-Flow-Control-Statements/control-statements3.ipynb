```python
#Strong number Program
n=int(input("Enter a number :"))
num=n #Saving n
sum=0
if n==0:
   print("Enter a number greater than 0")
else:
   while n!=0:
      #it gives last number
      last =n%10 
      #Finding factorial of last number 
      prod=1
      for i in range(1,last+1):
         prod=prod*i
      sum=sum+prod
      n=n//10
   #checking if the input number sum is equal to the number
   if sum==num:
      print(f"{num} is a strong number")
   else:
      print(f"{num} is not a strong number")
```

```python
#Finding power of a number
num=int(input("Enter a number :"))
power=int(input("Enter the power :"))
prod=1
for i in range(1,num+1):
   prod=prod*power
print(f"{num} to the power {power} is {prod}")
```

```python
#Kids Finding Armstrong number concept
#sum of cubes of digits of a number is equal to the number itself
n=int(input("Enter a number :"))
num=n #saving n
sum=0
while n!=0:
   last =n%10
   #Finding cube of last number
   cube=last**3
   sum=sum+cube
   #Updating n
   n=n//10
#Comparing sum to num   
if sum==num:
   print(f"{num} is a armstrong number")
else:
   print(f"{num} is not a armstrong number")
```

```python
#Orginal concept of armstrong number
n=int(input("Enter a number :"))
a=b=n
count=0
while n!=0:
   n = n//10
   count=count+1
#now counter contains how many digits are there in the input number
sum=0
while a!=0:
   last=a%10
   sum=sum+last**count
   a=a//10
if sum==b:
   print(f"{b} is a armstrong number")
else:
   print(f"{b} is not a armstrong number")
```

## Continue

## Skip the remaining portion of curret iteration & continue with next iteration

```python
for i in range(1,11):
   print("Pankaj")
   if i%3==0:
      continue
   print(i)
```

## break

## Whenever break is encounter loop terminate we came out of loop

```python
for i in range(1,11):
   if i%3==0:
      break
   print(i)
```

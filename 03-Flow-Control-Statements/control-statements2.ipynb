```python
num=int(input("Enter a number :"))
prod=1
i=1
while i<=num:
    prod=prod*i
    i+=1
print(f"Factorial of {num} is {prod}")
```

```python
#Find how many digits are there in any given input number
num=int(input("Enter any number :"))
count=0
if num==0:
    count=1
else:
    while num!=0:
        num=num//10
        count+=1    
print(f"There are {count} digits are there in your given number")
```

```python
#Find a factor of a number 
n=int(input("Enter a number :"))
for i in range(1,n+1):
    if n%i==0:
        print(f"Factors of {n} are {i}:")
```

```python
#Storing factors in a list
n=int(input("Enter a number :"))
l=[] #Empty list
for i in range(1,n+1):
    if n%i==0:
        l.append(i)
print(f"Factors of {n} are {l}")
```

```python
#Perfect Number n is called as a perfect number if the sum of all factors of n excluding itself is  equal to n
n=int(input("Enter a number :"))
sum=0
for i in range(1,n):
    if n%i==0:
        sum=sum+i
if sum==n:
    print(f"{n} is a perfect number")
else:
    print(f"{n} is not a perfect number")
```

```python
#Sum of digits of a number
n=int(input('Enter a number :'))
num=n
sum=0
while n!=0:
    last=n%10
    sum=sum+last
    n=n//10
print(f"sum of digits of {num} is {sum}")
```

```python
#Find maximum among 3 distict numbers
a=int(input("Enter first number : "))
b=int(input("Enter second number : "))
c=int(input("Enter third number : "))
d=  a if (a>b) and (a>c) else b if (b>c) else c
print(f"The largest among {a},{b},& {c} is {d}") 
```

# Iterative Statements

## If we want to execute some code any number of time we use iterative statements

## There are 2 types of iterative statements 
## 1. for loop
## 2. while loop

## Syntax for for loop : for x in sequence:
## code

## sequence can be list, tuple,striing,range,dictionary

```python
for x in "Akash":
   print(x)
```

```python
i=0
for x in "Akash":
   print(f"The element at index {i} is {x}")
   i+=1
```

```python
for i in range(5):
   print("Akash")
```

```python
for i in range(1,6):
   print("Akash")
```

```python
for i in range(1,101):
   if i%2==0:
      print("Even number is ",i)
```

```python
for i in range(10,1,-1): ##for decreasing loop it will 
   print(i)
```

```python
#Find product of first 5 natural number using loop
prod=1
for i in range(1,6):
   prod=prod*i
   
print("Product of 5 natural number is : ",prod)
```

```python
#Program to find any factorial of any number 
num=int(input("Enter a number : "))
prod=1
for i in range(1,num+1):
   prod=prod*i
print("Factorial of",num,"is",prod)
```

```python
#Given a list of integer write a program to find the sum of all elements in the list 
l=[1,3,8,2,10]
sum=0
for i in l:
   sum=sum+i
print("The sum of all elements is :",sum)
```

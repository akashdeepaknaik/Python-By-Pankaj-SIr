#### Join( )

```python
## join jodta hai list of strings ko
l=['Hello', 'i', 'am', 'Akash']
a=' '.join(l)
print(a)
print(type(a))
```

```python
l=['Hello', 'i', 'am', 'Akash']
a='@'.join(l)
print(a)
print(type(a))
```

```python
l=['Hello', 'i', 'am', 'Akash']
a='#'.join(l)
print(a)
print(type(a))
```

```python
l=['Hello', 'i', 'am', 'Akash']
a='*'.join(l)
print(a)
print(type(a))
```

```python
#to print even length words in the string
s="Hello i am akash"
for word in s.split():
    if len(word)%2==0:
        print(word)
```

```python
#Akash Deepak Nail ==> ADN
#Deepak Atmaram Naik ==> DAN
name=input()
a=name.split()
updated=[]
for word in a[ : len(a)-1]: #iterate over all but not last
    updated.append(word[0]) #no validation
updated.append(a[len(a)-1]) #last word append
print(' '.join(updated))
```

##### find(  )

```python
## It will return index of first occurence of the given strings/substring if it is not present then it returns -1
S="Pankaj sir is Pankaj Sir"
print(S.find("Pankaj"))
```

```python
### There is another version of find() in which we can tell from where to start
# s.(substring , start inde , end index) search sub in between to endindex-1 & return index of first index occurence if it is present otherwise returns -1
s="pankaj sir se ek hi aadmi bacha skta hai ohh hai khud pankaj sir"
print(s.find("pankaj",6,15))
```

##### Slicing

```python
# Syntax [start:end:step]
# if we did not specify starting index by default python start from 0
a='Akash'
print(a[:5])
# if we did not specify end then printing will goes till end of string
print(a[0:])
#if we did not specify any start or end value by default it will print the entire string
print(a[:])
```

```python
#Strory of a slice operator
'''
in forward direction 
default begin with 0
default end is len(string)
default step is 1

in backward direction
default start is -1
default end is -(leng(string)+1)
default step is -1
'''
```

```python
name="Akash"
print(name[4:2:-1]) #-1 indicates direction is reverse start is 4 and end is (2+1) ie, 3
```

```python
print(name[1:4:-1]) # No such string exists so empty
```

```python
#For reversing a string in python
print(name[::-1])
```

```python
#Multiply operator for string it is repeting strings
print(name*2) 
#addition operator is concatinates strings
print(name+name)
```

```python
name="Akash"
for i in name[::-1]:
    print(i)
```

```python
# palindrome name alli  _jahaj kina modlu or jahaj aadmele en madidvi space kottirtive so edanna remove madlike python alli three way ede
# First one rstrip removes space from right of the string
name="Akash    "
updated=name.rstrip()
print(updated)
print(len(updated))
```

```python
# Second one lstrip remove spaces from left of the string
name= "    Akash"
updated=name.lstrip()
print(updated)
print(len(updated))
```

```python
# Third one strip for removing both side spaces
name="    Akash     "
updated=name.strip()
print(updated)
print(len(updated))
```

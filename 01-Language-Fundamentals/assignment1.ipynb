# Q2. Take two numbers as input from the user and print their sum, difference,
# product, and quotient

```python
a=float(input("Enter a number : "))
b=float(input("Enter second number : "))

print(f"Sum of {a} & {b} is {a+b}")
print(f"Difference of {a} & {b} is {b-a}")
print(f"Product of {a} & {b} is {a*b}")
print(f"Quotient of {a} & {b} is {a%b}")
```

# Q7 Ask the user for a temperature in Celsius (string input). Convert it to ,  float
# then calculate and print temperature in Fahrenheit.
# Conversion formula: FahrenheitTemp = (CelsiusTemp ∗ (9/5)) + 32

```python
temp=input("Enter your temperature in celsius : ")

FahrenheitTemp = (float(temp) * (9/5)) + 32
print(f"Temperature in Fahrenheit is : {int(FahrenheitTemp)}")
```

# Q8. Take the radius (r) as user input and print the area.

# Q9. Ask the user for: Principal (P), Rate (R), Time (T). Convert all to float and
# compute simple interest: SI = (P ∗ R ∗ T)/100


```python
principal=int(input("Enter principal amount : "))
rate=int(input("Enter the rate amount : "))
time=int(input("Enter the time required : "))
simple_interest=(principal*rate*time)/100
print(f"Simple interest is : {float(simple_interest)}")
```

# Q10. Take a decimal number as input (like 45.78 ) and output its:
# • integer part - 45
# • fractional part - .78

```python
num=float(input("Enter a decimal number : "))
integer_num=int(num)
fract=num-integer_num
print(f"Integer part is : {integer_num}")
print(f"Fractional Part is : {fract}")
```

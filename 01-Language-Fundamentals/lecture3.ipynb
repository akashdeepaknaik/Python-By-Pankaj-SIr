```python
t.append(2)
```

## Range()

```python
range(5) #it will generate values from 0 to 4 
```

```python
for element in range(5):print(element) #its a loop
```

```python
#format 2 (start & end)
range(5,10)
```

```python
for element in range(5,10):print(element)
```

```python
#format 3 (start,end,step)
for element in range(2,20,2):print(element)
```

```python
for i in range(2,20,2):print(i)
```

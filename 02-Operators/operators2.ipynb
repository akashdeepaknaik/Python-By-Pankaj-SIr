# Ternary Operator

## first value if (condition) else second value

```python
a=10
b=20
c=100 if a<b else 1000
print(c)
```

```python
#Largest of two numbers
a=int(input("Enter a number : "))
b=int(input("Enter Second number : "))
c= a if (a>b) else b
print(f"{c} is larger")
```

```python
#Largest among 3 numbers
a=int(input("Enter a number : "))
b=int(input("Enter second number : "))
c=int(input("Enter third number : "))
#a is largest if this is true a>b and a>c
d=  a if (a>b) and (a>c) else b if (b>c) else c
print(f"{d} is larger among 3 numbers")
```
